## <u> Amazon Sales Data </u>

Q1. What is the average rating for each product category? 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("amazon.csv")
df.head()
```

```python
#covert rating_count into numeric
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

avg_rating= df.groupby("category")["rating"].mean().sort_values(ascending=False)
print(avg_rating)
```

Q2. What are the top rating_count products by category? 

```python
top_rating_product= df.sort_values("rating_count",ascending=False)
top_5=top_rating_product[["category","product_name","rating","rating_count"]].head(5)
print(top_5)
```

Q3. What is the distribution of discounted prices vs. actual prices? 

```python
plt.figure(figsize=(8,5))
plt.hist(df["actual_price"],alpha=0.5,label="Actual Price")
plt.hist(df["discounted_price"],alpha=0.5 , label="Discounted Price")
plt.legend()
plt.title("Price Distribution")
plt.show()
```

Q4.How does the average discount percentage vary across categories? 

```python
# remove % and convert to numeric
df["discount_percentage"]=pd.to_numeric(df["discount_percentage"].str.replace("%",""))

avg_discount = df.groupby("category")["discount_percentage"].mean().sort_values(ascending=False)
print(avg_discount)
```

Q5.What are the most popular product names? 

```python
df["product_name"].value_counts().head(10)
```

Q6.What are the most popular product keywords? 

```python
import re
from collections import Counter
#combine all product name into one text
text=" ".join(df["product_name"].dropna()).lower()
#remove special characters
text = re.sub(r"[^a-zA-Z0-9\s]","",text)

# split into words
words= text.split()
stopwords = {"the","and","for","with","to","of","in","a","on","usb","cable"}
words = [word for word in words if word not in stopwords and len(word)> 2]

#count frequency
word_count = Counter(words)
print(word_count.most_common(20))
```

Q7.What are the most popular product reviews? 

```python

df["rating_count"] = pd.to_numeric(
    df["rating_count"].astype(str).str.replace(",", ""),
    errors="coerce"
)


top_reviews = df.sort_values(by="rating_count", ascending=False).head(10)

print(top_reviews[["product_name", "rating", "rating_count", "review_title"]])
```

Q8.What is the correlation between discounted_price and rating? 

```python
df["discounted_price"] = (
    df["discounted_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df = df.dropna(subset=["discounted_price", "rating"])

correlation = df["discounted_price"].corr(df["rating"])

print("Correlation:", correlation)
```

Q9.What are the Top 5 categories based on the highest ratings? 

```python
# convert rating to numeric
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")


df = df.dropna(subset=["rating"])


top5_categories = (
    df.groupby("category")["rating"]
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

print(top5_categories)
```

## <u> Spotify Sales Data</u>

Q1.Load the dataframe and ensure data quality by checking for missing values and 
duplicate rows. Handle missing values and remove duplicate rows if necessary. 

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("spotify.csv")

# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Check for duplicate rows
duplicates = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicates)

# Remove duplicate rows
df = df.drop_duplicates()

# Handle missing values
# Drop rows with any missing values
df = df.dropna()

print("\nAfter cleaning:")
print("Missing values:\n", df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())


print("\nCleaned data preview:")
print(df.head())
```

Q2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a 
histogram. 

```python


print(df.columns)

# convert to numeric
df["Popularity"] = pd.to_numeric(df["Popularity"], errors="coerce")

# Drop missing values
popularity = df["Popularity"].dropna()

# Create histogram
plt.figure()
plt.hist(popularity, bins=20)
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.title("Distribution of Track Popularity")
plt.show()


print(popularity.describe())
```

Q4.Which artist has the highest number of tracks in the dataset? Display the count of tracks 
for each artist using a countplot. 

```python
import seaborn as sns


# Count number of tracks per artist
artist_counts = df['Artist'].value_counts()

print(artist_counts)

top_artist = artist_counts.idxmax()
top_count = artist_counts.max()

print("Artist with highest number of tracks:", top_artist)
print("Number of tracks:", top_count)

plt.figure(figsize=(12,6))

top_10 = df['Artist'].value_counts().head(10)

sns.countplot(
    y='Artist',
    data=df,
    order=top_10.index
)

plt.title("Top 10 Artists by Number of Tracks")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.show()
```

Q7.For the top 5 most popular artists, what are their most popular tracks? List the track 
name for each artist. 

```python
top_artists_max = df.groupby('Artist')['Popularity'].max().sort_values(ascending=False).head(5)

# Step 2: Retrieve the track name for
results = []
for artist, max_pop in top_artists_max.items():
   
    track_name = df[(df['Artist'] == artist) & (df['Popularity'] == max_pop)]['Track Name'].iloc[0]
    results.append({'Artist': artist, 'Track Name': track_name, 'Popularity': max_pop})

top_5_artists_tracks = pd.DataFrame(results)

print(top_5_artists_tracks)
```
